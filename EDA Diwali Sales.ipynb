# Exploratory Data Analysis

## Gender

```python
plt.figure(figsize=(6,4))
ax=sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

#### Inference:
    Females display a higher propensity for purchasing as compared to males.

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data = sales_gen)
```

#### Inference:
Purchase data indicates that females have a higher average number of transactions than males.

## Age

```python
ax = sns.countplot(data=df, x ='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs age group 
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)


sns.barplot(x='Age Group',y='Amount', data=sales_age)
```

#### Inference:
Females remain dominant shoppers in this age group. Their shopping frequency surpasses males, specifically in the age group of 26-35.

## State

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10))

sales_orders = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)


sns.barplot(data=sales_orders, x='State', y='Orders', ax=ax1)
ax1.set_title('Total Number of Orders from Top 10 States')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')


sales_amount = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)



sns.barplot(data=sales_amount, x='State', y='Amount', ax=ax2)
ax2.set_title('Total Amount from Top 10 States')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')


plt.tight_layout()

plt.show()
```

#### Inference:
Unexpectedly Uttar Pradesh holds the most orders and amount. There is a positive relation between order and amount.

## Marital Status

```python
ax=sns.countplot(data=df,x='Marital_Status')

sns.set(rc={'figure.figsize':(7,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(6,5)})

sns.barplot(data=sales_state,x='Marital_Status',y='Amount',hue='Gender')
```

#### Inference:
Most of the buyers who are married distinctively married women have made higer purchases.

## Occupation

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 5))


ax1 = sns.countplot(data=df, x='Occupation', ax=axs[0])
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')
ax1.set_xlabel('Occupation')
ax1.set_ylabel('Count')
ax1.set_title('Occupation Count')


for bars in ax1.containers:
    ax1.bar_label(bars)

    
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax2 = sns.barplot(data=sales_state, x='Occupation', y='Amount', ax=axs[1], color='orange')
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right')
ax2.set_ylabel('Amount')
ax2.set_title('Total Amount by Occupation')


plt.tight_layout()


plt.show()
```

### Inference:
Most of the buyers are working in IT, Aviation and Healthcare sectors.

## Product Ctaegory

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 7))


ax1 = sns.countplot(data=df, x='Product_Category', ax=axs[0])
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')
ax1.set_xlabel('Product Category')
ax1.set_ylabel('Count')
ax1.set_title('Product Category Count')


for bars in ax1.containers:
    ax1.bar_label(bars)

    
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax2 = sns.barplot(data=sales_state, x='Product_Category', y='Amount', ax=axs[1], color='orange')
ax2.set_ylabel('Amount')
ax2.set_title('Total Amount by Product Category')


plt.tight_layout()

plt.show()
```

#### Inference:
From above graphs it is clearly visible that the highest sold products are from Food, Footwear and Electronics category. 

```python
sales_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})

sns.barplot(data=sales_state,x='Product_ID',y='Orders')
```
